```python
import tensorflow as tf
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
```

```python
(train_img, train_labels), (test_img, test_labels) = fashion_mnist.load_data() # Train and test data are numpy arrays
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_img.shape # Training feature space
28*28
```

```python
nn = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape = (28,28)), tf.keras.layers.Dense(256, activation='relu'), tf.keras.layers.Dense(10)])
```

Layers in a keras are stored in a list, whereas in PyTorch they are all inputs to __init__

Also Collab has a search Stack Overflow feature!!

```python
nn.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics = ['accuracy'])
nn.fit(train_img, train_labels, epochs = 12) # Plateus at a certain point
```

```python
# A second neural network learning basic pattern
import numpy as np
```

```python
input_feature_1 = np.random.normal(size = (1,500))
input_feature_2 = np.random.normal(size = (1,500))
train = np.append(input_feature_1, input_feature_2, axis = 0)
train.shape
```

```python
input_feature_1[0][1:5] # some inputs
```

```python
outputs = 0.2 * input_feature_1 + 0.6 * input_feature_2 + 27
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(1)])
```

```python
model.compile(loss = tf.keras.losses.MSE)
model.fit(train.reshape(500,2), outputs.reshape(500,1), epochs = 1500)
```
